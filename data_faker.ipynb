```python
from faker import Faker
from faker.providers import BaseProvider
import random
import pymysql
import uuid
import string
```

# Initialize the DB Connection and Faker Class

```python
connection = pymysql.connect(
    host="localhost",
    user="root",
    password="root",
    database="rumah_sakit_database",
    charset="utf8mb4",
    cursorclass=pymysql.cursors.DictCursor
)
```

```python
fake = Faker()
```

# Constant of Any Value for TINYINT

```python
constant_tiny_int = [1, 2, 3, 4]
```

# Generate Our Own Boolean Probability when There's Faker Boolean

```python
def generate_boolean(probability_of_true):
    return random.random() < probability_of_true
```

# Generate Custom Hospital Name

```python
hospital_name = fake.company()
class CustomHospitalProvider(BaseProvider):
    def hospital_name(self):
        return self.generator.company_suffix() + " " + self.generator.city()+ " " + self.generator.word(ext_word_list=["Hospital", "Medical Center", "Clinic"])
```

# Generate Fake org_tbl_hospital

```python

def generate_fake_hospital_data(num_hospital):
    hospitals = []

    for _ in range(num_hospital):
        fake.add_provider(CustomHospitalProvider)
        fake_hospital_name = fake.hospital_name()
        print(fake_hospital_name)
        
        id = uuid.uuid4()
        hospital_name = fake_hospital_name
        created_at = fake.date_time_between(start_date="-30d", end_date="now")
        updated_at = fake.date_time_between(start_date="-29d", end_date="now")
        deleted_at = None

        hospitals.append((id, hospital_name, created_at, updated_at, deleted_at))
    
    return hospitals 
                    
def insert_fake_hospital(hospital):
    with connection.cursor() as cursor:
        sql = "INSERT INTO org_tbl_hospital(id, hospital_name, created_at, updated_at, deleted_at) VALUES (%s, %s, %s, %s, %s)"
        cursor.executemany(sql, hospital)
    connection.commit()


# Generate Fake Hospital and insert to DB
fake_hospitals = generate_fake_hospital_data(1000)
insert_fake_hospital(fake_hospitals)

connection.close
```

# Create Faker Data for Medical Order
